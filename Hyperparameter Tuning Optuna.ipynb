```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math

# ═══════════════════════════════════════════════════════════════════════════════
# 1.  EFFICIENT MULTI-SCALE ATTENTION (EMA)
# ═══════════════════════════════════════════════════════════════════════════════

class EMA(nn.Module):
    """Efficient Multi-Scale Attention — plug-in attention module untuk neck YOLO."""

    def __init__(self, channels: int, factor: int = 8):
        super().__init__()
        self.groups = factor
        assert channels // self.groups > 0, "channels harus kelipatan factor"
        self.softmax  = nn.Softmax(dim=-1)
        self.agp      = nn.AdaptiveAvgPool2d((1, 1))
        self.pool_h   = nn.AdaptiveAvgPool2d((None, 1))
        self.pool_w   = nn.AdaptiveAvgPool2d((1, None))
        self.gn       = nn.GroupNorm(channels // self.groups, channels // self.groups)
        self.conv1x1  = nn.Conv2d(channels // self.groups, channels // self.groups,
                                   kernel_size=1, bias=False)
        self.conv3x3  = nn.Conv2d(channels // self.groups, channels // self.groups,
                                   kernel_size=3, padding=1, bias=False)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        b, c, h, w = x.size()
        group_x = x.reshape(b * self.groups, -1, h, w)
        x_h = self.pool_h(group_x)
        x_w = self.pool_w(group_x).permute(0, 1, 3, 2)
        hw  = self.conv1x1(torch.cat([x_h, x_w], dim=2))
        x_h, x_w = torch.split(hw, [h, w], dim=2)
        x_w = x_w.permute(0, 1, 3, 2)
        a1  = torch.sigmoid(self.gn(x_h))
        a2  = torch.sigmoid(self.gn(x_w))
        group_x = group_x * a1.expand_as(group_x) * a2.expand_as(group_x)
        x1  = self.agp(group_x)
        x2  = self.conv3x3(group_x).reshape(b * self.groups, -1, 1)
        x11 = self.softmax(x1.reshape(b * self.groups, -1, 1))
        x12 = x1.reshape(b * self.groups, -1, 1)
        x21 = self.softmax(x2)
        x22 = x2
        weights = (torch.matmul(x11.transpose(1,2), x12) +
                   torch.matmul(x21.transpose(1,2), x22)).reshape(b * self.groups, 1, h, w)
        out = group_x * weights.sigmoid().expand_as(group_x)
        return out.reshape(b, c, h, w)


# ═══════════════════════════════════════════════════════════════════════════════
# 2.  POWERFUL IoU (PIoU) LOSS
# ═══════════════════════════════════════════════════════════════════════════════

def piou_loss(pred: torch.Tensor, target: torch.Tensor,
              alpha: float = 1.0, beta: float = 1.0) -> torch.Tensor:
    """
    PIoU loss untuk axis-aligned bounding boxes.
    pred, target: (N, 4) format (x1, y1, x2, y2)
    """
    inter_x1 = torch.max(pred[:, 0], target[:, 0])
    inter_y1 = torch.max(pred[:, 1], target[:, 1])
    inter_x2 = torch.min(pred[:, 2], target[:, 2])
    inter_y2 = torch.min(pred[:, 3], target[:, 3])
    inter_w  = (inter_x2 - inter_x1).clamp(min=0)
    inter_h  = (inter_y2 - inter_y1).clamp(min=0)
    inter    = inter_w * inter_h
    area_p   = (pred[:, 2] - pred[:, 0]) * (pred[:, 3] - pred[:, 1])
    area_t   = (target[:, 2] - target[:, 0]) * (target[:, 3] - target[:, 1])
    union    = area_p + area_t - inter + 1e-7
    iou      = inter / union

    enc_x1   = torch.min(pred[:, 0], target[:, 0])
    enc_y1   = torch.min(pred[:, 1], target[:, 1])
    enc_x2   = torch.max(pred[:, 2], target[:, 2])
    enc_y2   = torch.max(pred[:, 3], target[:, 3])
    c2       = (enc_x2 - enc_x1) ** 2 + (enc_y2 - enc_y1) ** 2 + 1e-7

    p_cx = (pred[:, 0] + pred[:, 2]) / 2
    p_cy = (pred[:, 1] + pred[:, 3]) / 2
    t_cx = (target[:, 0] + target[:, 2]) / 2
    t_cy = (target[:, 1] + target[:, 3]) / 2
    rho2 = (p_cx - t_cx) ** 2 + (p_cy - t_cy) ** 2

    corner_dist = (
        (pred[:, 0] - target[:, 0]) ** 2 + (pred[:, 1] - target[:, 1]) ** 2 +
        (pred[:, 2] - target[:, 2]) ** 2 + (pred[:, 1] - target[:, 1]) ** 2 +
        (pred[:, 0] - target[:, 0]) ** 2 + (pred[:, 3] - target[:, 3]) ** 2 +
        (pred[:, 2] - target[:, 2]) ** 2 + (pred[:, 3] - target[:, 3]) ** 2
    )

    piou = iou - rho2 / c2 - alpha * corner_dist / (c2 + 1e-7) * beta
    return (1 - piou).unsqueeze(-1)


# ═══════════════════════════════════════════════════════════════════════════════
# 3.  PATCH ultralytics — PIoU loss + EMA injection
# ═══════════════════════════════════════════════════════════════════════════════

from ultralytics import YOLO
from ultralytics.nn.modules import C2f
from ultralytics.nn.modules.block import SPPF
import ultralytics.utils.loss as ult_loss
from ultralytics.utils.metrics import bbox_iou
from ultralytics.utils.tal import bbox2dist

if not hasattr(ult_loss.BboxLoss, '_truly_original_forward'):
    ult_loss.BboxLoss._truly_original_forward = ult_loss.BboxLoss.forward

def _piou_bbox_loss_forward(self, pred_dist, pred_bboxes, anchor_points,
                             target_bboxes, target_scores, target_scores_sum, fg_mask,
                             imgsz=None, stride=None):
    """Override BboxLoss.forward dengan PIoU. Kompatibel Ultralytics 8.4.x."""
    weight = target_scores.sum(-1)[fg_mask].unsqueeze(-1)
    iou = bbox_iou(pred_bboxes[fg_mask], target_bboxes[fg_mask], xywh=False, CIoU=True)

    piou_val = piou_loss(pred_bboxes[fg_mask], target_bboxes[fg_mask])
    loss_iou = ((1.0 - iou) * 0.3 + piou_val * 0.7) * weight
    loss_iou = loss_iou.sum() / target_scores_sum

    if self.dfl_loss is not None:
        target_ltrb = bbox2dist(anchor_points, target_bboxes, self.dfl_loss.reg_max - 1)
        loss_dfl = self.dfl_loss(
            pred_dist[fg_mask].view(-1, self.dfl_loss.reg_max),
            target_ltrb[fg_mask]
        ) * weight
        loss_dfl = loss_dfl.sum() / target_scores_sum
    else:
        if imgsz is not None and stride is not None:
            tl = bbox2dist(anchor_points, target_bboxes) * stride
            tl[..., 0::2] /= imgsz[1]
            tl[..., 1::2] /= imgsz[0]
            pd = pred_dist * stride
            pd[..., 0::2] /= imgsz[1]
            pd[..., 1::2] /= imgsz[0]
            loss_dfl = (F.l1_loss(pd[fg_mask], tl[fg_mask], reduction='none')
                        .mean(-1, keepdim=True) * weight)
            loss_dfl = loss_dfl.sum() / target_scores_sum
        else:
            loss_dfl = torch.tensor(0., device=pred_dist.device)
    return loss_iou, loss_dfl

ult_loss.BboxLoss.forward = _piou_bbox_loss_forward
print('✓ PIoU berhasil di-patch ke BboxLoss (Ultralytics 8.4.x kompatibel)')


class C2fWithEMA(nn.Module):
    """C2f + EMA attention wrapper untuk neck YOLOv8."""

    def __init__(self, c2f_module: C2f):
        super().__init__()
        self.c2f = c2f_module
        ch = c2f_module.cv2.conv.out_channels
        factor = 8
        while ch % factor != 0 and factor > 1:
            factor //= 2
        self.ema = EMA(channels=ch, factor=factor)

    def forward(self, x):
        return self.ema(self.c2f(x))


def inject_ema_to_neck(model):
    """
    Deteksi akhir backbone secara otomatis via SPPF layer,
    lalu ganti semua C2f di neck dengan C2fWithEMA.
    Kompatibel dengan YOLOv8 n/s/m/l/x.
    """
    backbone_end = None
    for i, layer in enumerate(model.model.model):
        if isinstance(layer, SPPF):
            backbone_end = i
            break
    neck_start = (backbone_end or 9) + 1

    injected = 0
    for i, layer in enumerate(model.model.model):
        if i >= neck_start and isinstance(layer, C2f):
            model.model.model[i] = C2fWithEMA(layer)
            injected += 1
    print(f'✓ EMA disematkan ke {injected} C2f layer di neck (backbone_end={backbone_end}, neck_start={neck_start})')
    return model


# ═══════════════════════════════════════════════════════════════════════════════
# 4.  FOCAL LOSS GAMMA PATCH
#     Ultralytics 8.4.x tidak menerima fl_gamma sebagai argumen model.train().
#     Kita implementasikan via monkey-patch pada v8DetectionLoss.__init__
#     agar mengganti self.bce dengan versi focal-loss-aware.
# ═══════════════════════════════════════════════════════════════════════════════

_FL_GAMMA = 0.0

def set_fl_gamma(gamma: float):
    """Set global focal loss gamma sebelum memanggil model.train()."""
    global _FL_GAMMA
    _FL_GAMMA = gamma


class FocalBCEWithLogitsLoss(nn.Module):
    """BCEWithLogitsLoss dengan focal loss modulation."""
    def __init__(self, gamma=0.0, reduction='none'):
        super().__init__()
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, pred, target):
        bce = F.binary_cross_entropy_with_logits(pred, target, reduction='none')
        if self.gamma > 0:
            p_t = torch.exp(-bce)
            bce = bce * ((1 - p_t) ** self.gamma)
        if self.reduction == 'mean':
            return bce.mean()
        elif self.reduction == 'sum':
            return bce.sum()
        return bce


if not hasattr(ult_loss.v8DetectionLoss, '_truly_original_init'):
    ult_loss.v8DetectionLoss._truly_original_init = ult_loss.v8DetectionLoss.__init__

def _patched_v8det_init(self, *args, **kwargs):
    ult_loss.v8DetectionLoss._truly_original_init(self, *args, **kwargs)
    if _FL_GAMMA > 0:
        self.bce = FocalBCEWithLogitsLoss(gamma=_FL_GAMMA, reduction='none')

ult_loss.v8DetectionLoss.__init__ = _patched_v8det_init
print('✓ Focal Loss Gamma patch siap (set via set_fl_gamma() sebelum training)')


print('\n✓ Modul EMA, PIoU & Focal Loss siap.')
```

```python
import optuna
import wandb
import json
from ultralytics import YOLO
from pathlib import Path
import shutil, os, gc, torch

N_TRIALS      = 55
TUNE_EPOCHS   = 8 
TUNE_IMG_SIZE = 640
TUNE_DEVICE   = DEVICE
TUNE_DIR      = OUTPUT_DIR / 'optuna_trials_v5'
TUNE_DIR.mkdir(exist_ok=True)

OPTUNA_DB = TUNE_DIR / 'optuna_study.db'
OPTUNA_STORAGE = f'sqlite:///{OPTUNA_DB.as_posix()}'
BEST_PARAMS_JSON = TUNE_DIR / 'best_params.json'

# ── Shortcut: jika best_params.json sudah ada, langsung pakai ────────────────
FORCE_RETUNE = True   # Set False setelah re-tune selesai

_TUNING_SKIP = False
if BEST_PARAMS_JSON.exists() and not FORCE_RETUNE:
    try:
        with open(BEST_PARAMS_JSON) as f:
            _cached_bp = json.load(f)
        if _cached_bp:
            best_params = _cached_bp
            _TUNING_SKIP = True
            print(f'✓ best_params.json sudah tersedia — skip Optuna tuning.')
            print(f'\n{"="*50}')
            print(f' Best Hyperparameters (dari cache)')
            print(f'{"="*50}')
            for k, v in best_params.items():
                print(f'  {k:20s}: {v}')
    except Exception:
        pass
else:
    _TUNING_SKIP = False
    if FORCE_RETUNE:
        if OPTUNA_DB.exists():
            OPTUNA_DB.unlink()
        if BEST_PARAMS_JSON.exists():
            BEST_PARAMS_JSON.unlink()
        print('FORCE_RETUNE=True — DB dan cache lama dihapus, tuning dimulai dari awal.')


def objective(trial: optuna.Trial) -> float:
    """Objective Optuna: 0.6 * mAP50-95 + 0.4 * recall."""

    optimizer_name = trial.suggest_categorical('optimizer', ['SGD', 'AdamW'])
    lr0            = trial.suggest_float('lr0',          1e-4,  2e-2, log=True)
    lrf            = trial.suggest_float('lrf',          0.01,  0.2,  log=True)
    momentum       = trial.suggest_float('momentum',     0.85,  0.97)
    weight_decay   = trial.suggest_float('weight_decay', 1e-4,  5e-3, log=True)
    batch          = trial.suggest_categorical('batch',  [8, 16])
    box            = trial.suggest_float('box',          5.0,   10.0)
    cls            = trial.suggest_float('cls',          0.3,   2.0)
    dfl            = trial.suggest_float('dfl',          0.8,   2.5)
    fl_gamma       = trial.suggest_float('fl_gamma',     0.5,   2.5)
    warmup_epochs  = trial.suggest_float('warmup_epochs', 2.0,  5.0)
    cos_lr         = trial.suggest_categorical('cos_lr', [True, False])
    close_mosaic   = trial.suggest_int('close_mosaic',   10,    30)
    mixup          = trial.suggest_float('mixup',        0.0,   0.3)

    trial_name = f'trial_{trial.number:03d}'

    run = wandb.init(
        project = WANDB_PROJECT,
        entity  = WANDB_ENTITY,
        name    = trial_name,
        group   = 'optuna_tuning_v5',
        config  = {
            'optimizer': optimizer_name,
            'lr0': lr0, 'lrf': lrf, 'weight_decay': weight_decay,
            'momentum': momentum, 'batch': batch,
            'box': box, 'cls': cls, 'dfl': dfl,
            'fl_gamma': fl_gamma, 'warmup_epochs': warmup_epochs,
            'cos_lr': cos_lr, 'close_mosaic': close_mosaic, 'mixup': mixup,
            'epochs': TUNE_EPOCHS, 'imgsz': TUNE_IMG_SIZE,
        },
        reinit  = True,
    )

    try:
        set_fl_gamma(fl_gamma)

        model = YOLO('yolov8n.pt')
        inject_ema_to_neck(model)

        results = model.train(
            data          = str(DATA_YAML),
            epochs        = TUNE_EPOCHS,
            imgsz         = TUNE_IMG_SIZE,
            batch         = batch,
            optimizer     = optimizer_name,
            lr0           = lr0,
            lrf           = lrf,
            momentum      = momentum,
            weight_decay  = weight_decay,
            box           = box,
            cls           = cls,
            dfl           = dfl,
            warmup_epochs = warmup_epochs,
            cos_lr        = cos_lr,
            close_mosaic  = close_mosaic,
            mixup         = mixup,
            device        = TUNE_DEVICE,
            project       = str(TUNE_DIR),
            name          = trial_name,
            exist_ok      = True,
            verbose       = False,
            workers       = 4,
            patience      = 5,
            amp           = True,
            save          = True,
            plots         = False,
            deterministic = True,
            seed          = 42,
        )

        val_metrics = results.results_dict
        map50_95 = float(val_metrics.get('metrics/mAP50-95(B)', 0.0))
        map50    = float(val_metrics.get('metrics/mAP50(B)', 0.0))
        recall   = float(val_metrics.get('metrics/recall(B)', 0.0))
        prec     = float(val_metrics.get('metrics/precision(B)', 0.0))

        score = 0.60 * map50_95 + 0.40 * recall

        wandb.log({
            'mAP50': map50, 'mAP50_95': map50_95,
            'recall': recall, 'precision': prec,
            'composite_score': score,
        })
        run.finish()

        del model
        gc.collect()
        torch.cuda.empty_cache()

        return score

    except Exception as e:
        print(f'  [Trial {trial.number}] Error: {e}')
        run.finish()
        gc.collect()
        torch.cuda.empty_cache()
        return 0.0


if not _TUNING_SKIP:
    optuna.logging.set_verbosity(optuna.logging.WARNING)

    STUDY_NAME = 'yolov8_progressive_tuning'

    if OPTUNA_DB.exists():
        try:
            _tmp_study = optuna.load_study(study_name=STUDY_NAME, storage=OPTUNA_STORAGE)
            _complete = [t for t in _tmp_study.trials
                         if t.state == optuna.trial.TrialState.COMPLETE and t.value is not None and t.value > 0]
            if len(_tmp_study.trials) > 0 and len(_complete) == 0:
                print(f'DB lama berisi {len(_tmp_study.trials)} trial tapi SEMUA gagal (score=0). Menghapus DB...')
                del _tmp_study
                OPTUNA_DB.unlink()
                print('DB lama dihapus, tuning akan dimulai dari awal.')
            else:
                del _tmp_study
        except Exception as e:
            print(f'Gagal load DB lama ({e}), menghapus dan mulai ulang...')
            OPTUNA_DB.unlink(missing_ok=True)

    study = optuna.create_study(
        direction  = 'maximize',
        study_name = STUDY_NAME,
        storage    = OPTUNA_STORAGE,
        load_if_exists = True,
        sampler    = optuna.samplers.TPESampler(seed=42),
    )

    completed_trials = [t for t in study.trials
                        if t.state == optuna.trial.TrialState.COMPLETE and t.value is not None and t.value > 0]
    remaining = N_TRIALS - len(completed_trials)

    print(f'Trial COMPLETE (score>0): {len(completed_trials)}/{N_TRIALS}')
    all_states = {}
    for t in study.trials:
        s = t.state.name
        all_states[s] = all_states.get(s, 0) + 1
    print(f'Status semua trial di DB: {all_states}')

    if remaining > 0:
        print(f'Menjalankan {remaining} trial lagi...')
        study.optimize(objective, n_trials=remaining, show_progress_bar=True)
    else:
        print(f'Semua {N_TRIALS} trial sudah selesai. Skip tuning.')

    completed_trials = [t for t in study.trials
                        if t.state == optuna.trial.TrialState.COMPLETE and t.value is not None and t.value > 0]

    if len(completed_trials) == 0:
        all_trials = study.trials
        completed_any = [t for t in all_trials if t.state == optuna.trial.TrialState.COMPLETE]
        print(f'\nTotal trials di DB: {len(all_trials)}, COMPLETE (any): {len(completed_any)}')
        if completed_any:
            print('Semua trial COMPLETE tapi score <= 0. Mengambil trial dengan score tertinggi...')
            best_trial = max(completed_any, key=lambda t: t.value if t.value is not None else float('-inf'))
        else:
            raise RuntimeError('Tidak ada trial yang berhasil! Periksa error di atas dan jalankan ulang cell ini.')
    else:
        best_trial = max(completed_trials, key=lambda t: t.value)

    best_params = best_trial.params
    print(f'\n{"="*50}')
    print(f' Best Hyperparameters (trial {best_trial.number}, score={best_trial.value:.4f})')
    print(f'{"="*50}')
    for k, v in best_params.items():
        print(f'  {k:20s}: {v}')

    with open(BEST_PARAMS_JSON, 'w') as f:
        json.dump(best_params, f, indent=2)
    print(f'\nbest_params tersimpan di {BEST_PARAMS_JSON}')
```

## Cell 4b — Recovery: Rebuild best_params dari Trial Folders

Jalankan cell ini jika tuning terputus dan Optuna DB tidak tersedia.
Akan scan semua `trial_xxx/args.yaml` + `weights/best.pt` dan memilih yang terbaik.

```python
from pathlib import Path
import yaml, json
from ultralytics import YOLO

TUNE_DIR_SCAN = OUTPUT_DIR / 'optuna_trials_v5'
BEST_PARAMS_JSON = TUNE_DIR_SCAN / 'best_params.json'

PARAM_KEYS_REQUIRED = [
    'optimizer', 'lr0', 'lrf', 'weight_decay', 'momentum',
    'batch', 'box', 'cls', 'dfl',
    'warmup_epochs', 'cos_lr',
]

PARAM_KEYS_OPTIONAL = {
    'fl_gamma': 0.0,
}


def get_trial_params(args_path: Path):
    if not args_path.exists():
        return None
    with open(args_path) as f:
        args = yaml.safe_load(f)
    if not args:
        return None
    params = {}
    for k in PARAM_KEYS_REQUIRED:
        if k not in args:
            return None
        v = args[k]
        if k == 'batch':
            params[k] = int(round(v))
        elif k in ('optimizer', 'cos_lr'):
            params[k] = v
        else:
            params[k] = float(v)
    for k, default in PARAM_KEYS_OPTIONAL.items():
        params[k] = float(args[k]) if k in args else default
    return params


def get_composite_score(weights_path: Path, data_yaml: str, device) -> dict:
    """Validasi best.pt dan return composite score + metrics."""
    model = YOLO(str(weights_path))
    metrics = model.val(data=data_yaml, batch=16, device=device, verbose=False)
    rd = metrics.results_dict
    map50_95 = float(rd.get('metrics/mAP50-95(B)', 0.0))
    recall   = float(rd.get('metrics/recall(B)', 0.0))
    score    = 0.60 * map50_95 + 0.40 * recall
    return {'map50_95': map50_95, 'recall': recall, 'score': score}


trial_dirs = sorted([d for d in TUNE_DIR_SCAN.iterdir()
                     if d.is_dir() and d.name.startswith('trial_')])
print(f'Ditemukan {len(trial_dirs)} folder trial di {TUNE_DIR_SCAN}')

results = []
for td in trial_dirs:
    args_path = td / 'args.yaml'
    best_pt   = td / 'weights' / 'best.pt'
    params    = get_trial_params(args_path)
    if params is None:
        print(f'  Skip {td.name}: args.yaml tidak lengkap')
        continue
    if not best_pt.exists():
        print(f'  Skip {td.name}: best.pt tidak ada')
        continue
    try:
        m = get_composite_score(best_pt, str(DATA_YAML), DEVICE)
        results.append((td.name, params, m))
        print(f'  {td.name}: score={m["score"]:.4f}  mAP50-95={m["map50_95"]:.4f}  recall={m["recall"]:.4f}')
    except Exception as e:
        print(f'  Skip {td.name}: {e}')

if not results:
    if BEST_PARAMS_JSON.exists():
        with open(BEST_PARAMS_JSON) as f:
            best_params = json.load(f)
        if best_params:
            print(f'\nRecovery dari folder gagal, tapi best_params.json tersedia:')
            for k, v in best_params.items():
                print(f'  {k:20s}: {v}')
        else:
            print('\nTidak ada trial yang bisa dipakai dan best_params.json kosong.')
            best_params = {}
    else:
        print('\nTidak ada trial yang bisa dipakai.')
        best_params = {}
else:
    best_name, best_params, best_metrics = max(results, key=lambda x: x[2]['score'])
    with open(BEST_PARAMS_JSON, 'w') as f:
        json.dump(best_params, f, indent=2)
    print(f'\n═══ Best dari {len(results)} trial (recovery) ═══')
    print(f'  Trial terbaik : {best_name}')
    print(f'  Score         : {best_metrics["score"]:.4f}')
    print(f'  mAP50-95      : {best_metrics["map50_95"]:.4f}')
    print(f'  Recall        : {best_metrics["recall"]:.4f}')
    for k, v in best_params.items():
        print(f'  {k:20s}: {v}')
    print(f'\n✓ best_params tersimpan di {BEST_PARAMS_JSON}')
```
